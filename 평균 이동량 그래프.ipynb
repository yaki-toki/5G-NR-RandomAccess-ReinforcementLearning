```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load the CSV file
file_path = './202310281824/q-learning/50000_q_learning_new_rewards_proposed.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
```

```python
data.describe()
```

```python
# Load the CSV file
file_path = './202310281824/ddpg/50000_ddpg_rewards_proposed.csv'
ddpg_data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
ddpg_data.head()
```

```python
ddpg_data.describe()
```

```python
# Setting the style and font size for the plots
sns.set(style="whitegrid")
plt.rcParams['font.size'] = 30
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['axes.labelsize'] = 30
plt.rcParams['axes.titlesize'] = 30
plt.rcParams['xtick.labelsize'] = 30
plt.rcParams['ytick.labelsize'] = 30
plt.rcParams['legend.fontsize'] = 30
plt.rcParams['figure.titlesize'] = 30
```

```python
# Calculate the moving average of the rewards
window_size = 50
moving_avg_rewards = data['Reward'].rolling(window=window_size).mean()

# Adjusting the plot for better visualization and readability
plt.figure(figsize=(16, 9))

# Plotting Q-Learning: Raw Rewards and Moving Average (Up to 4000 episodes)
plt.plot(data['Reward'][:4000], label='Q-Learning Raw rewards', color='orange', alpha=0.35, linewidth=2)
plt.plot(moving_avg_rewards[:4000], label='Q-Learning Moving Average', color='red', linewidth=2.5)

# Adding titles and labels
# plt.title("Comparison of Q-Learning and DDPG Rewards (Up to 4000 Episodes)")
plt.xlabel("Episode")
plt.ylabel("Raw rewards")
plt.legend(loc='upper right')  
plt.tight_layout()

plt.show()
```

```python
# Calculate the moving average of the rewards
window_size = 50
ddpg_moving_avg_rewards = ddpg_data['Reward'].rolling(window=window_size).mean()

# Adjusting the plot for better visualization and readability
plt.figure(figsize=(16, 9))

# Plotting DDPG: Raw Rewards and Moving Average (Up to 4000 episodes)
plt.plot(ddpg_data['Reward'][:4000], label='DDPG Raw rewards', alpha=0.35, linewidth=2)
plt.plot(ddpg_moving_avg_rewards[:4000], label='DDPG Moving Average', color='green', linewidth=2.5)

# Adding titles and labels
# plt.title("Comparison of Q-Learning and DDPG Rewards (Up to 4000 Episodes)")
plt.xlabel("Episode")
plt.ylabel("Raw rewards")
plt.legend(loc='upper right')  
plt.tight_layout()

plt.show()
```
